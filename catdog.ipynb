```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# Load the data
with open('X_gray.pickle', 'rb') as f:
    X = pickle.load(f)

with open('y.pickle', 'rb') as f:
    y = pickle.load(f)

# X = np.load('X_gray.npy')
# y = np.load('y.npy')

# Preprocessing the data

if len(X.shape) > 2:
    X_flat = X.reshape(X.shape[0], -1)
else:
    X_flat = X

# Spliting the data into training and testing sets
X_train, X_test, y_train, y_test, X_train_orig, X_test_orig = train_test_split(X_flat, y, X, test_size=0.2, random_state=42)

# Standardizing the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Training the SVM classifier
clf = svm.SVC(kernel='linear')  
clf.fit(X_train, y_train)

# Making the predictions
y_pred = clf.predict(X_test)

# Defining class names
class_names = {0: 'cat', 1: 'dog'}  

# Evaluating the classifier
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=[class_names[0], class_names[1]])

print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:")
print(report)

# Displaying the sample predictions with images
print("\nSample Predictions with Images:")
num_samples = 10
fig, axes = plt.subplots(2, num_samples // 2, figsize=(15, 5))
axes = axes.ravel()

for i in range(num_samples):
    axes[i].imshow(X_test_orig[i], cmap='gray')
    axes[i].set_title(f"Pred: {class_names[y_pred[i]]}\nActual: {class_names[y_test[i]]}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
